```python
import pandas
df=pandas.DataFrame([[1,2,3],[4,5,6]],columns=["price","values","range"],index=["first","sec"])
print(df)
print("Mean********\n",df.mean())
print("mean of all:",df.mean().mean())
print("Type of mean:",type(df.mean()))
print("price mean is :",df.price.mean())
print("Max price :",df.price.max())
```
